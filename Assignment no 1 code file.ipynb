```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.width", 1000)

file_path = "updated dataset.csv"
try:
    df = pd.read_csv(file_path)
    print("âœ… Data loaded successfully")
except Exception as e:
    print(f"âŒ Error loading data: {e}")
    exit()

print("\nðŸ”¹ First 5 Rows:")
print(df.head())

print("\nðŸ”¹ Dataset Shape:", df.shape)

print("\nðŸ”¹ Data Types:")
print(df.dtypes)

print("\nðŸ”¹ Null Values in Each Column:")
print(df.isnull().sum())

print("\nðŸ”¹ Statistical Summary:")
print(df.describe(include='all'))

print("\nðŸ“Š Generating Visualizations...")

df.hist(figsize=(12, 10))
plt.suptitle("Histogram of Numerical Columns")
plt.show()

if 'gender' in df.columns and 'diabetes' in df.columns:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x='gender', hue='diabetes')
    plt.title("Gender vs Diabetes Count")
    plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

print("\nâœ… EDA Completed Successfully!")
```
